```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from mpl_toolkits.mplot3d import Axes3D
```

```python
# import your data
df= pd.read_csv("ml_data_salary1.csv")
df.head()
```

```python
x = df[['Age', 'YearsExperience']]
y = df['Salary'] 
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# Creat and fit Your model
model= LinearRegression()
model.fit(x, y)
```

```python
# Predictions
df["prediction"]= model.predict(x)
```

```python
# print coefficient
print('Intercept:',model.intercept_)
print("Coefficient:", model.coef_)
```

```python
fig= plt.figure(figsize=(8,6))
ax= fig.add_subplot(111, projection="3d")

ax.scatter(x_test['Age'], x_test['YearsExperience'], y_test, color='blue', label='Actual')
ax.scatter(x_test['Age'], x_test['YearsExperience'], model.predict(x_test), color='green', label='Predicted')

ax.set_xlabel('Age')
ax.set_ylabel('YearsExperience')
ax.set_zlabel('Salary')
ax.set_title('Multiple Regression: Age & Experience vs Salary')
ax.legend()
plt.show()
```
